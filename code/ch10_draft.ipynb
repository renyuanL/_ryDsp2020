```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl


N=100
n= np.arange(N)
x0= np.cos(n*.1)
#x1= np.random.random(N)*.1
x1= np.cos(n)*.1
x2= x0+x1

```

```python
pl.plot(n,x0,'r',n,x1,'g',n,x2,'b')
#pl.plot(n,x2)
```

```python
h0= np.array([1,1,1,1,1,1,1,1])/8
h1= np.array([1,-1,1,-1,1,-1,1,-1])/8
y0= np.convolve(x2,h0,'same')
y1= np.convolve(x2,h1,'same')

pl.plot(n,x2,'r',n,y0,'g',n,y1,'b')
```

```python
H0= np.fft.fft(h0,N)
H1= np.fft.fft(h1,N)
k=  np.arange(H0.size)

pl.plot(k,abs(H0),'r',k,abs(H1),'g')
pl.grid(True)
```

```python
X2= np.fft.fft(x2)
Y0= X2*H0
Y1= X2*H1
pl.plot(k,abs(Y0),'r',k,abs(Y1),'g')
pl.grid(True)
```
